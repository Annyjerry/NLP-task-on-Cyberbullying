```python
# Removing white spaces from the data
def remove_whiteSpaces(Text):
    if isinstance(Text, str):
        text = re.sub('[^a-zA-Z0-9]', ' ', Text)
        text = re.sub(r'\s+', ' ', Text)
        return Text
    return Text

df['text'] = df['text'].apply(lambda x: remove_whiteSpaces(x))

df.sample(10)
```

```python
# Removing special character
def remove_s_characters(Text):
    str_text = str(Text)
    return re.sub(r'[^\w\s]', '', str_text)


df['text'] = df['text'].apply(lambda x: remove_s_characters(x))

df.sample(10)
```

```python
#Remove punctuations, links, stopwords, mentions and \r\n new line characters
def strip_all_entities(text):
    if isinstance(text, str):
        text = text.replace('\r', '').replace('\n', ' ').lower()  # Remove \n and \r and lowercase
        text = re.sub(r"(?:\@|https?\://)\S+", "", text)  # Remove links and mentions
        text = re.sub(r'[^\x00-\x7f]', r'', text)  # Remove non utf8/ascii characters
    return text

df['text'] = df['text'].apply(lambda x: strip_all_entities(x))

df.sample(10)
```

```python
#remove contractions
def decontract(text):
    if isinstance(text, str):
        text = re.sub(r"can\'t", "can not", text)
        text = re.sub(r"n\'t", " not", text)
        text = re.sub(r"\'re", " are", text)
        # Add more contraction replacements as needed
    return text

df['text'] = df['text'].apply(lambda x: decontract(x))
df.sample(10)
```

```python
#clean hashtags at the end of the sentence, and keep those in the middle of the sentence by removing just the "#" symbol
def clean_hashtags(tweet):
    if isinstance(tweet, str):
        # Remove last hashtags
        new_tweet = " ".join(word.strip() for word in re.split('#(?!(?:hashtag)\\b)[\\w-]+(?=(?:\\s+#[\\w-]+)*\\s*$)', tweet))
        # Remove hashtag symbols from words in the middle of the sentence
        new_tweet2 = " ".join(word.strip() for word in re.split('#|_', new_tweet))
        return new_tweet2
    return tweet

df['text'] = df['text'].apply(lambda x: clean_hashtags(x))
df.sample(10)
```

```python
#Stemming
import nltk
from nltk.stem import PorterStemmer

def stemmer(text):
    if isinstance(text, str):
        tokenized = nltk.word_tokenize(text)
        ps = PorterStemmer()
        return ' '.join([ps.stem(word) for word in tokenized])
    return text

df['text'] = df['text'].apply(lambda x: stemmer(x))
df.sample(10)
```

```python
#Lemmatization 
def lemmatize(text):
    tokenized = nltk.word_tokenize(text)
    lm = WordNetLemmatizer()
    return ' '.join([lm.lemmatize(words) for words in tokenized])

df['text'] = df['text'].apply(lambda x: lemmatize(x))
df.sample(10)
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.pipeline import Pipeline


tfidf = TfidfTransformer()
clf = CountVectorizer()

X_cv =  clf.fit_transform(df['text'])

tf_transformer = TfidfTransformer(use_idf=True).fit(X_cv)
X_tf = tf_transformer.transform(X_cv)
```
